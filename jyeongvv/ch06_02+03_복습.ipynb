```python
#@title 합성

customers = ['Alice', 'Ben', 'Charles', 'Don', 'Ein']
ids = [f"1{i+1:03d}"for i in range(len(customers))]
ids, customers
```

```python
df1 = pd.DataFrame({
    'id': ids,
    'name': customers,
})
df1
```

```python
df2 = pd.DataFrame({
    "id": np.random.choice([f"1{i+1:03d}" for i in range(len(customers)+2)], 10, replace=True)
})
df2
```

```python
df2['money'] = df2.id.apply(lambda x: int(x) * abs(np.random.randn()) * 1000).apply(int)
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how='inner')
```

```python
pd.merge(df1, df2, how='outer')
```

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

```python
pd.merge(df1, df2, on='id')
```

```python
df3 = df1.copy()
df3.columns = ['id', 'data']
df4 = df2.copy()
df4.columns = ['id', 'data']
```

```python
df4
```

```python
pd.merge(df3, df4, on='id')
```

```python
pd.merge(df3, df4, on='id', suffixes=['_A', '_B'])
```
