```python
def draw_line(style):
    plt.title('나이와 재산의 상관관계')
    plt.plot(
        range(10, 50, 10), # x축 데이터
        [1, 4, 9, 16], # y축 데이터
        style
    )
    # plt.xlabel('x축라벨')
    plt.xlabel('나이')
    # plt.ylabel('y축라벨')
    plt.ylabel('재산')
    plt.show()
```

### 그리드 설정 (grid)


```python
# 보조선 활성화/비활성화 : plt.grid(T/F)
def draw_line(grid=False, style='g-'):
    plt.title('나이와 재산의 상관관계')
    plt.plot(
        range(10, 50, 10), # x축 데이터
        [1, 4, 9, 16], # y축 데이터
        style
    )
    # plt.xlabel('x축라벨')
    plt.xlabel('나이')
    plt.grid(grid)
    # plt.ylabel('y축라벨')
    plt.ylabel('재산')
    plt.show()
```

```python
draw_line(grid=True) # 기본값은 False
draw_line(grid=False)
```

### 여러 개의 선 그어보기

```python
t = np.arange(0., 5., 0.2)
t
```

```python
# 1. 여러 개의 선 그리기 (하나의 plot 사용)
plt.title('라인 플롯에서 여러개의 선')
plt.plot(
    t, t, 'r--',
    t, t ** 2, 'go:',
    t, t ** 3, 'bs-.',
)
plt.show()
```

```python
# 2. 하나의 plot이 아니라 여러 개 -> 겹치는.
plt.title('여러개의 plot 명령으로 한 그림에 겹쳐서 표현')
plt.plot((np.arange(4)+1) ** 2)
plt.plot((np.arange(4)-1) ** 3)
plt.show()
```

```python
plt.plot((np.arange(4)+1) ** 2)
plt.show()
plt.plot((np.arange(4)-1) ** 3)
plt.show()
```

```python
# 여러 개의 플롯이 겹쳤을 경우 라벨 표시
X = np.linspace(-np.pi, np.pi, 256)
X
```

```python
# 코사인, 사인 곡선
C, S = np.cos(X), np.sin(X)
C, S
```

```python
plt.title('삼각함수 plot')
plt.plot(X, C)
plt.plot(X, S)
plt.show()
```

```python
plt.title('삼각함수 plot')
plt.plot(X, C, label='cosine')
plt.plot(X, S, label='sine')
plt.legend(loc=2)
plt.show()
```

#### `subplot`
* plot 명령을 여러 번 실행 => 같은 그래프 내부에 여러 직선 표시
* subplot : 행과 열 단위로 묶어서 여러 데이터를 병행해서 표시

```python
import numpy as np
```

```python
# 위아래 행 배치
# 행, 열, 서브플롯의 위치
plt.subplot(2, 1, 1)
plt.plot(np.arange(5) ** 2)
plt.title('차트1')
plt.subplot(2, 1, 2)
plt.plot(np.arange(10) ** 3, 'r--')
plt.title('차트2')
# 겹치는 문제
plt.tight_layout() # 안겹치게 레이아웃 조정
plt.show()
```

```python
# 양옆 열 배치
# 행, 열, 서브플롯의 위치
# 행, 열, 서브플롯의 위치
plt.subplot(1, 2, 1)
plt.plot(np.arange(5) ** 2)
plt.title('차트1')
plt.subplot(1, 2, 2)
plt.plot(np.arange(10) ** 3, 'r--')
plt.title('차트2')
# 겹치는 문제
plt.tight_layout() # 안겹치게 레이아웃 조정
plt.show()
```

```python
# 2x2 플롯 배치
plt.figure(figsize=(10,6))
plt.subplot(2, 2, 1)
plt.plot(np.arange(5) ** 2)
plt.title('차트1')
plt.subplot(2, 2, 2)
plt.plot(np.arange(10) ** 3, 'g.-')
plt.title('차트2')
plt.subplot(2, 2, 3)
plt.plot(np.arange(5) ** 2)
plt.title('차트1')
plt.subplot(2, 2, 4)
plt.plot(np.arange(10) ** 3, 'y:')
plt.title('차트2')
# 겹치는 문제
plt.tight_layout() # 안겹치게 레이아웃 조정
plt.show()
```

### 다양한 차트 활용
* `lineplot` 선 차트
* 그외에 여러 가지 차트가 존재

```python
y = [2, 3, 1]
x = np.arange(3)
xlabel = ['가', '나', '다'] # len(x)와 같아야함
plt.title('세로 bar')
plt.xticks(x, xlabel) # x축의 개별 값에다 이름 붙여줌
plt.yticks(sorted(y)) # y축의 개별 값에다 이름 붙여줌
plt.bar(x, y)
plt.show()
```

```python
y = [2, 3, 1]
x = np.arange(3)
xlabel = ['가', '나', '다'] # len(x)와 같아야함
plt.title('가로 bar')
plt.yticks(x, xlabel) # x축의 개별 값에다 이름 붙여줌
plt.xticks(sorted(y)) # y축의 개별 값에다 이름 붙여줌
plt.barh(x, y)
plt.show()
```

#### 파이 차트
* 원형의 파이 모양 그래프

```python
labels = ['호박파이', '사과파이', '호두파이', '무화과파이']
sizes = [15, 30, 23, 50]
colors = ['brown', 'skyblue', 'pink', 'lightgreen']
explode = (0, 0.1, 0.2, 0.4)
plt.title('파이차트')
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

#### 히스토그램 (Histogram)

```python
x = np.random.randn(1000)
plt.title('히스토그램')
arrays, bins, patches = plt.hist(x, bins=10)
plt.show()
```

#### 스캐터 플롯 (산점도)

* x과 y 값 사이의 상관성이 있는 변수들
* 두 변수 사이의 상관관계를 점들의 밀집 정도를 통해서 확인하는 그래프

```python
X = np.random.normal(0, 1, 100)
Y = np.random.normal(0, 1, 100)
plt.title('산점도')
plt.scatter(X, Y)
plt.show()
```
